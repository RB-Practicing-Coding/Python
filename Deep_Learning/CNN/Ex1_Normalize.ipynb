Steps to build a neural network for diabetes detection:
 1. Loaddataset
 2. Split data for training and testing
 3. Normalize the feature values
 4. Build neural network model
 5. Train the neural network
 6. Evaluate the trained neural network

```python
# Step 1: Load dataset
# Load the mnist dataset using keras
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)

# Shown sample images
# Show one example in training set
idx = 150
x = X_train[idx, :]
print
y = y_train[idx]
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(np.reshape(x,(28, 28)), cmap=plt.cm.gray)
plt.title('Corresponding label: %i\n' %y, fontsize=20)
```

```python
# Step 2. Split data for training and testing
# Step 2 is not necessary since the dataset was divided into training and test set
# Show the number of training images and test images
print("Number of training images: ", len(y_train))
print('Number of test images: ', len(y_test))

# Step 3. Normalize the feature values
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
# Step 4. Build neural network model
from sklearn.neural_network import MLPClassifier

# Size of input layer = 8
# Number of hidden layers = 1
# Size of hidden layer = 2
# Size of output layer = 1
model = MLPClassifier(
    hidden_layer_sizes=(2),
    activation='relu',
    learning_rate_init=0.001,
    max_iter=2000,
    random_state=2,
    verbose=True)
```

```python
# Step 5. Train the designed model
model.fit(X_train, y_train)
```

```python
# Show the weights and biases of the trained network
print('Weights = ', model.coefs_)
print('Biases = ', model.intercepts_)
print('Size of output layer: ', model.n_outputs_)
print('Activation function of output layer: ', model.out_activation_)

# Show the plot of the loss curve
loss_values = model.loss_curve_
plt.plot(loss_values)
plt.xlabel('Iteration')
plt.ylabel('Cost value')
plt.show()
```

```python
# Step 6. Evaluate the trained neural network
from sklearn.metrics import accuracy_score

# Compute the accuracy of the trained network on the training set
predict_train = model.predict(X_train)
accuracy_train = accuracy_score(y_train, predict_train)
print('Accuracy of the trained network on the training set: %.2f %%' % (100 * accuracy_train))

# Compute the accuracy of the train network on the test set
predict_test = model.predict(X_test)
accuracy_test = accuracy_score(y_test, predict_test)
print('Accuracy of the trained network on the test set: %.2f %%' % (100 * accuracy_test))
```

```python
# Change hidden layer to (64, 64, 32)

# Size of input layer = 8
# Number of hidden layers = 1
# Size of hidden layer = 3, the first is 64 neurons, the second is 64 and the third is 32
# Size of output layer = 1
model = MLPClassifier(
    hidden_layer_sizes=(64, 64, 32),
    activation='relu',
    learning_rate_init=0.001,
    max_iter=2000,
    random_state=2,
    verbose=True)

# Step 5. Train the designed model
model.fit(X_train, y_train)

# Show the weights and biases of the trained network
print('Weights = ', model.coefs_)
print('Biases = ', model.intercepts_)
print('Size of output layer: ', model.n_outputs_)
print('Activation function of output layer: ', model.out_activation_)

# Show the plot of the loss curve
loss_values = model.loss_curve_
plt.plot(loss_values)
plt.xlabel('Iteration')
plt.ylabel('Cost value')
plt.show()

# Step 6. Evaluate the trained neural network
from sklearn.metrics import accuracy_score

# Compute the accuracy of the trained network on the training set
predict_train = model.predict(X_train)
accuracy_train = accuracy_score(y_train, predict_train)
print('Accuracy of the trained network on the training set: %.2f %%' % (100 * accuracy_train))

# Compute the accuracy of the train network on the test set
predict_test = model.predict(X_test)
accuracy_test = accuracy_score(y_test, predict_test)
print('Accuracy of the trained network on the test set: %.2f %%' % (100 * accuracy_test))
```
