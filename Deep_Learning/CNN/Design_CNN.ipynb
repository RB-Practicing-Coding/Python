```python
# prompt: import library to design a cnn
import matplotlib.pyplot as plt
import numpy as np
import pathlib
import os

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
# prompt: Design a cnn model name LeNet 5 - 1998, use activation relu
rows = 32
cols = 32
dim = 1

model = Sequential([
    layers.InputLayer(input_shape=(rows, cols, dim)), # Define input shape explicitly
    # Feature scaling using Rescaling from keras.layers.preprocessing
    # layers.preprocessing.Rescaling(scale=1./255, input_shape=(rows, cols, dim)),
    # Conv: 6 filters 5x5
    layers.Conv2D(6, (5, 5), activation='relu'), # Added padding
    # Average pooling
    layers.AveragePooling2D(pool_size=(2, 2)),
    # Conv: 16 filters 5x5
    layers.Conv2D(16, (5, 5), activation='relu'), # Added padding option
    # Average pooling 2x2
    layers.AveragePooling2D(pool_size=(2, 2)),
    # Convert to feature vector
    layers.Flatten(),
    # FC 120 neurons
    layers.Dense(120, activation='relu'),
    # FC 84 neurons
    layers.Dense(84, activation='relu'),
    # Output layer: 10 neurons
    layers.Dense(10, activation='softmax')
])

model.summary()
```
